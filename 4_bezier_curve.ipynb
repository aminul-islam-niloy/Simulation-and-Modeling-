```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cubic_bezier(t, p0, p1, p2, p3):
    p0, p1, p2, p3 = np.array(p0), np.array(p1), np.array(p2), np.array(p3)
    return (1-t)**3 * p0 + 3*t*(1-t)**2 * p1 + 3*t**2*(1-t) * p2 + t**3 * p3
```

```python
def plot_cubic_bezier_curve(p0, p1, p2, p3):
   
    t_values = np.linspace(0, 1, 100)
    curve_points = [cubic_bezier(t, p0, p1, p2, p3) for t in t_values]

    plt.plot(*zip(*curve_points), label="Cubic Bezier Curve")
    plt.scatter(*zip(p0, p1, p2, p3), color='red', label="Control Points")
    plt.legend()
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Cubic Bezier Curve")
    plt.grid(True)
    plt.show()
```

```python
# Example control points
P0 = (0, 0)
P1 = (1, 3)
P2 = (4, -2)
P3 = (6, 1)
```

```python
# Plot the Cubic Bezier curve
plot_cubic_bezier_curve(P0, P1, P2, P3)
```
