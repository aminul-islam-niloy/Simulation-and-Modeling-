```python
def chemical_reactor(A, B, C, k1, k2, end_time, dt): 
    time = 0
    substance_A = []
    substance_B = []
    substance_C = []
    time_points = []

    while (time <= end_time):
        substance_A.append(A)
        substance_B.append(B)
        substance_C.append(C)
        time_points.append(time)
        time += dt

        A += (k2 * C - k1 * A * B) * dt
        B += (k2 * C - k1 * A * B) * dt
        C += (2 * k1 * A * B - 2 * k2 * C) * dt

    return time_points, substance_A, substance_B, substance_C
```

```python
def main():
    A, B, C, k1, k2, end_time, dt = 100, 50, 0, 0.008, 0.002, 10, 0.1
    time_points, substance_A, substance_B, substance_C = chemical_reactor(A, B, C, k1, k2, end_time, dt)

    print(f"{'Time (s)':<10} {'Substance A':<15} {'Substance B':<15} {'Substance C'}")
    for t, a, b, c in zip(time_points, substance_A, substance_B, substance_C):
        print(f"{t:<10.2f} {a:<15.2f} {b:<15.2f} {c}")
```

```python
if __name__ == "__main__":
    main()
```
